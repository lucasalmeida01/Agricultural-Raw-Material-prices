# Python insights - Agricultural Raw Material prices (1990-2020)

```python
#importando as bibliotecas que serão utilizadas no projeto
import pandas as pd
import numpy as np
```

```python
#criando nosso dataframe
df = pd.read_csv('agricultural_raw_material.csv')
df
```

### Traduzindo o nome das nossas colunas

```python
# Dicionário de traduções
traducoes = {
    'Month': 'Mês',
    'Coarse wool Price': 'Preço da lã grossa',
    'Coarse wool price % Change': 'Variação % do preço da lã grossa',
    'Copra Price': 'Preço do copra',
    'Copra price % Change': 'Variação % do preço do copra',
    'Cotton Price': 'Preço do algodão',
    'Cotton price % Change': 'Variação % do preço do algodão',
    'Fine wool Price': 'Preço da lã fina',
    'Fine wool price % Change': 'Variação % do preço da lã fina',
    'Hard log Price': 'Preço do tronco duro',
    'Hard log price % Change': 'Variação % do preço do tronco duro',
    'Hard sawnwood Price': 'Preço da madeira serrada dura',
    'Hard sawnwood price % Change': 'Variação % do preço da madeira serrada dura',
    'Hide Price': 'Preço do couro',
    'Hide price % change': 'Variação % do preço do couro',
    'Plywood Price': 'Preço da madeira compensada',
    'Plywood price % Change': 'Variação % do preço da madeira compensada',
    'Rubber Price': 'Preço da borracha',
    'Rubber price % Change': 'Variação % do preço da borracha',
    'Softlog Price': 'Preço do tronco macio',
    'Softlog price % Change': 'Variação % do preço do tronco macio',
    'Soft sawnwood Price': 'Preço da madeira serrada macia',
    'Soft sawnwood price % Change': 'Variação % do preço da madeira serrada macia',
    'Wood pulp Price': 'Preço da polpa de madeira',
    'Wood pulp price % Change': 'Variação % do preço da polpa de madeira'
}

# Renomeando as colunas
df = df.rename(columns=traducoes)

df
```

### Tratando os dados

```python
#verificandoa as informçãoes do dataframe, nome das colunas e tipo
df.info()
```

```python
#verificando a quantidade de registos nulos por coluna
df.isnull().sum()
```

### Tratando valores nulos, incorretos e inválidos

```python
#regex - expressões regulares
#removendo todos os caracteres que consideramos invalidos
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('',np.nan)
df = df.replace('MAY90', np.nan)

#Apagando os valores nulos
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
#convertendo tipo de dados (data para float)
#como temos colunas precos e percentual, ambos podem ser numeros fracionados
# Identificando a coluna que não será convertida
col_to_exclude = 'Mês'

# Selecionando todas as colunas, exceto a que queremos excluir
cols_to_convert = df.columns.difference([col_to_exclude])

# Convertendo as colunas selecionadas para float
df[cols_to_convert] = df[cols_to_convert].astype(float)

print(df)
print(df.dtypes)  # Para verificar os tipos das colunas
```

```python
#trazer top5 colunas
df.head()

#vamos converter a coluna Month para formato data, pois vamos projetar em timeseries
#timeseries - gráfico utilizando data que mostra a evolução ao longo do tempo
```

```python
df.Mês = pd.to_datetime(df.Mês.str.upper(), format='%b%y', yearfirst=False)

df = df.set_index('Mês')

#convertendo a coluna Month de strings para objetos datetime e define essa coluna como o índice do df, facilitando operações baseadas em tempo
```

```python
df.info()
df.head()
```

### Iniciando a Análise Exploratória

```python
#definindo alguns padrões para os gráficos
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Mapa de calor
# Lista de matérias-primas
# Correlação entre as colunas, para saber se existe uma relação direta
dados_brutos = ['Preço da lã grossa', 'Preço do copra', 'Preço do algodão', 'Preço da lã fina', 
                'Preço do tronco duro', 'Preço da madeira serrada dura', 'Preço do couro', 
                'Preço da madeira compensada', 'Preço da borracha', 'Preço do tronco macio', 
                'Preço da madeira serrada macia', 'Preço da polpa de madeira']

# Obtendo a matriz de correlação
matriz_correlacao = df[dados_brutos].corr()

# Definindo o tamanho do gráfico
fig = plt.figure(figsize=(12, 9))

# Mascarando a parte superior do triângulo, já que a matriz é simétrica (repetitiva)
mascara = np.triu(np.ones_like(matriz_correlacao, dtype=bool))
sns.heatmap(matriz_correlacao, vmax=0.8, mask=mascara, square=True, annot=True)

plt.show()
```

```python
plt.figure(figsize=(30,15))

lista_mudanca = ['Variação % do preço do copra', 'Variação % do preço do tronco macio', 'Variação % do preço da borracha',
                 'Variação % do preço do algodão', 'Variação % do preço da lã grossa', 'Variação % do preço da lã fina',
                 'Variação % do preço do tronco duro', 'Variação % do preço da madeira serrada dura', 
                 'Variação % do preço do couro', 'Variação % do preço da madeira compensada', 
                 'Variação % do preço da madeira serrada macia', 'Variação % do preço da polpa de madeira']
                 
# Gerar uma matriz de correlação para todo o conjunto de dados
matriz_correlacao = df[lista_mudanca].corr()
sns.heatmap(matriz_correlacao, annot=True)

plt.show()
```

```python
# Análise detalhada em uma das matérias-primas - lã grossa
# Relação do preço com a variação em um gráfico evolutivo e verificar se existe relação na mesma matéria-prima

# Separar as colunas para gráficos separados
axes = df[["Preço da lã grossa", "Variação % do preço da lã grossa"]].plot(figsize=(11,9), subplots=True, linewidth=1)

# Definir subplots=False para trazer as duas colunas no mesmo gráfico
# No entanto, separaremos as colunas em dois gráficos diferentes para análise mais clara
```

### Respondendo perguntas de negócio

    - Descubra a variação normal do preço de cada matéria prima

```python
# Vamos usar um histograma para gerar um gráfico para cada matéria-prima
# Criamos uma lista contendo os nomes das colunas do df que representam as variações percentuais de preços de cada matéria-prima
changelist = ['Variação % do preço do copra', 'Variação % do preço do tronco macio', 'Variação % do preço da borracha',
             'Variação % do preço do algodão', 'Variação % do preço da lã grossa', 'Variação % do preço da lã fina',
             'Variação % do preço do tronco duro', 'Variação % do preço da madeira serrada dura',
             'Variação % do preço do couro', 'Variação % do preço da madeira compensada',
             'Variação % do preço da madeira serrada macia', 'Variação % do preço da polpa de madeira']

for i in range(len(changelist)):
    plt.figure(figsize=(8, 8))
    df[changelist[i]].hist(linewidth=1)
    plt.xlabel('% Mudança')
    plt.ylabel('Contagem')
    plt.legend([changelist[i]], loc='upper right')  # Legenda apenas para a coluna atual
    plt.title(f'Histograma de {changelist[i]}')  # Adicionar um título ao gráfico
    plt.show()  # Mostrar o gráfico

# Pensando em investimentos, o ideal seria comprar matéria-prima em que a variação de preço é menor
# A maioria das matérias-primas possui um percentual de mudança inferior a 5%

```

    - Encontrar a matéria-prima com menor preço ao longo dos anos

```python
plt.figure(figsize=(10, 10))
lista_materias = ['Preço do copra', 'Preço do tronco macio', 'Preço da borracha', 'Preço do algodão', 
                  'Preço da lã grossa', 'Preço da lã fina', 'Preço do tronco duro', 'Preço da madeira serrada dura',
                  'Preço do couro', 'Preço da madeira compensada', 'Preço da madeira serrada macia', 'Preço da polpa de madeira']

for i in range(len(lista_materias)):
    plt.subplot(4,3,i+1)
    plt.subplots_adjust(hspace=1 ,wspace=0.5)
    plt.title(lista_materias[i])
    plt.plot(df[lista_materias[i]])
    plt.xticks(rotation=90)

plt.suptitle("Comparação de preços de matérias-primas")

# Identificamos que algodão e borracha são as matérias-primas mais baratas e que menos sofreram variações
```

```python
# Comparando algodão e borracha
plt.figure(figsize=(10, 10))
plt.plot(df[['Preço do algodão','Preço da borracha']])
plt.title("Comparação de preços de matérias-primas")
plt.xlabel('Anos')
plt.ylabel('Preços')
plt.legend(['Preço do algodão','Preço da borracha'], loc='upper center', bbox_to_anchor=(1.2,1))
# O algodão possui o menor preço
```
