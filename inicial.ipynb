# Python insights - Agricultural Raw Material prices (1990-2020)

```python
#importando as bibliotecas que ser√£o utilizadas no projeto
import pandas as pd
import numpy as np
```

```python
#criando nosso dataframe
df = pd.read_csv('agricultural_raw_material.csv')
df
```

### Traduzindo o nome das nossas colunas

```python
# Dicion√°rio de tradu√ß√µes
traducoes = {
    'Month': 'M√™s',
    'Coarse wool Price': 'Pre√ßo da l√£ grossa',
    'Coarse wool price % Change': 'Varia√ß√£o % do pre√ßo da l√£ grossa',
    'Copra Price': 'Pre√ßo do copra',
    'Copra price % Change': 'Varia√ß√£o % do pre√ßo do copra',
    'Cotton Price': 'Pre√ßo do algod√£o',
    'Cotton price % Change': 'Varia√ß√£o % do pre√ßo do algod√£o',
    'Fine wool Price': 'Pre√ßo da l√£ fina',
    'Fine wool price % Change': 'Varia√ß√£o % do pre√ßo da l√£ fina',
    'Hard log Price': 'Pre√ßo do tronco duro',
    'Hard log price % Change': 'Varia√ß√£o % do pre√ßo do tronco duro',
    'Hard sawnwood Price': 'Pre√ßo da madeira serrada dura',
    'Hard sawnwood price % Change': 'Varia√ß√£o % do pre√ßo da madeira serrada dura',
    'Hide Price': 'Pre√ßo do couro',
    'Hide price % change': 'Varia√ß√£o % do pre√ßo do couro',
    'Plywood Price': 'Pre√ßo da madeira compensada',
    'Plywood price % Change': 'Varia√ß√£o % do pre√ßo da madeira compensada',
    'Rubber Price': 'Pre√ßo da borracha',
    'Rubber price % Change': 'Varia√ß√£o % do pre√ßo da borracha',
    'Softlog Price': 'Pre√ßo do tronco macio',
    'Softlog price % Change': 'Varia√ß√£o % do pre√ßo do tronco macio',
    'Soft sawnwood Price': 'Pre√ßo da madeira serrada macia',
    'Soft sawnwood price % Change': 'Varia√ß√£o % do pre√ßo da madeira serrada macia',
    'Wood pulp Price': 'Pre√ßo da polpa de madeira',
    'Wood pulp price % Change': 'Varia√ß√£o % do pre√ßo da polpa de madeira'
}

# Renomeando as colunas
df = df.rename(columns=traducoes)

df
```

### Tratando os dados

```python
#verificandoa as inform√ß√£oes do dataframe, nome das colunas e tipo
df.info()
```

```python
#verificando a quantidade de registos nulos por coluna
df.isnull().sum()
```

### Tratando valores nulos, incorretos e inv√°lidos

```python
#regex - express√µes regulares
#removendo todos os caracteres que consideramos invalidos
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('',np.nan)
df = df.replace('MAY90', np.nan)

#Apagando os valores nulos
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
#convertendo tipo de dados (data para float)
#como temos colunas precos e percentual, ambos podem ser numeros fracionados
# Identificando a coluna que n√£o ser√° convertida
col_to_exclude = 'M√™s'

# Selecionando todas as colunas, exceto a que queremos excluir
cols_to_convert = df.columns.difference([col_to_exclude])

# Convertendo as colunas selecionadas para float
df[cols_to_convert] = df[cols_to_convert].astype(float)

print(df)
print(df.dtypes)  # Para verificar os tipos das colunas
```

```python
#trazer top5 colunas
df.head()

#vamos converter a coluna Month para formato data, pois vamos projetar em timeseries
#timeseries - gr√°fico utilizando data que mostra a evolu√ß√£o ao longo do tempo
```

```python
df.M√™s = pd.to_datetime(df.M√™s.str.upper(), format='%b%y', yearfirst=False)

df = df.set_index('M√™s')

#convertendo a coluna Month de strings para objetos datetime e define essa coluna como o √≠ndice do df, facilitando opera√ß√µes baseadas em tempo
```

```python
df.info()
df.head()
```

### Iniciando a An√°lise Explorat√≥ria

```python
#definindo alguns padr√µes para os gr√°ficos
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Mapa de calor
# Lista de mat√©rias-primas
# Correla√ß√£o entre as colunas, para saber se existe uma rela√ß√£o direta
dados_brutos = ['Pre√ßo da l√£ grossa', 'Pre√ßo do copra', 'Pre√ßo do algod√£o', 'Pre√ßo da l√£ fina', 
                'Pre√ßo do tronco duro', 'Pre√ßo da madeira serrada dura', 'Pre√ßo do couro', 
                'Pre√ßo da madeira compensada', 'Pre√ßo da borracha', 'Pre√ßo do tronco macio', 
                'Pre√ßo da madeira serrada macia', 'Pre√ßo da polpa de madeira']

# Obtendo a matriz de correla√ß√£o
matriz_correlacao = df[dados_brutos].corr()

# Definindo o tamanho do gr√°fico
fig = plt.figure(figsize=(12, 9))

# Mascarando a parte superior do tri√¢ngulo, j√° que a matriz √© sim√©trica (repetitiva)
mascara = np.triu(np.ones_like(matriz_correlacao, dtype=bool))
sns.heatmap(matriz_correlacao, vmax=0.8, mask=mascara, square=True, annot=True)

plt.show()
```

```python
plt.figure(figsize=(30,15))

lista_mudanca = ['Varia√ß√£o % do pre√ßo do copra', 'Varia√ß√£o % do pre√ßo do tronco macio', 'Varia√ß√£o % do pre√ßo da borracha',
                 'Varia√ß√£o % do pre√ßo do algod√£o', 'Varia√ß√£o % do pre√ßo da l√£ grossa', 'Varia√ß√£o % do pre√ßo da l√£ fina',
                 'Varia√ß√£o % do pre√ßo do tronco duro', 'Varia√ß√£o % do pre√ßo da madeira serrada dura', 
                 'Varia√ß√£o % do pre√ßo do couro', 'Varia√ß√£o % do pre√ßo da madeira compensada', 
                 'Varia√ß√£o % do pre√ßo da madeira serrada macia', 'Varia√ß√£o % do pre√ßo da polpa de madeira']
                 
# Gerar uma matriz de correla√ß√£o para todo o conjunto de dados
matriz_correlacao = df[lista_mudanca].corr()
sns.heatmap(matriz_correlacao, annot=True)

plt.show()
```

### Respondendo perguntas de neg√≥cio

    - Descubra a varia√ß√£o normal do pre√ßo de cada mat√©ria prima

```python
# Vamos usar um histograma para gerar um gr√°fico para cada mat√©ria-prima
# Criamos uma lista contendo os nomes das colunas do df que representam as varia√ß√µes percentuais de pre√ßos de cada mat√©ria-prima
changelist = ['Varia√ß√£o % do pre√ßo do copra', 'Varia√ß√£o % do pre√ßo do tronco macio', 'Varia√ß√£o % do pre√ßo da borracha',
             'Varia√ß√£o % do pre√ßo do algod√£o', 'Varia√ß√£o % do pre√ßo da l√£ grossa', 'Varia√ß√£o % do pre√ßo da l√£ fina',
             'Varia√ß√£o % do pre√ßo do tronco duro', 'Varia√ß√£o % do pre√ßo da madeira serrada dura',
             'Varia√ß√£o % do pre√ßo do couro', 'Varia√ß√£o % do pre√ßo da madeira compensada',
             'Varia√ß√£o % do pre√ßo da madeira serrada macia', 'Varia√ß√£o % do pre√ßo da polpa de madeira']

for i in range(len(changelist)):
    plt.figure(figsize=(8, 8))
    df[changelist[i]].hist(linewidth=1)
    plt.xlabel('% Mudan√ßa')
    plt.ylabel('Contagem')
    plt.legend([changelist[i]], loc='upper right')  # Legenda apenas para a coluna atual
    plt.title(f'Histograma de {changelist[i]}')  # Adicionar um t√≠tulo ao gr√°fico
    plt.show()  # Mostrar o gr√°fico

# Pensando em investimentos, o ideal seria comprar mat√©ria-prima em que a varia√ß√£o de pre√ßo √© menor
# A maioria das mat√©rias-primas possui um percentual de mudan√ßa inferior a 5%

```
